```python
from os import listdir


def load_doc(filename):
    file = open(filename,'r')
    
    text = file.read()
    
    file.close()
    
    return text
```

```python
def process_docs(dir):
    
    for filename in listdir(dir):
        
        if not filename.endswith('.txt'):
            next
        directory = dir+'/'+filename
        load_doc(directory)
        print('Loaded %s' %filename )
```

```python
dir = 'review_polarity/txt_sentoken/neg'
process_docs(dir)
```

```python
# Split into tokens

filename = 'review_polarity/txt_sentoken/neg/cv853_29119.txt'
text = load_doc(filename)
tokens = text.split()
print(tokens)
```

```python
import string
import nltk

stop_words = nltk.corpus.stopwords.words('english')
def preprocessing_pipeline(tokens):
    tokens = tokens.split()
    preprocessed_text = [w for w in tokens if w not in string.punctuation] # Removing punctuation
    preprocessed_text = [w for w in preprocessed_text if w.isalpha()] # Removing everything except words
    preprocessed_text = [w.lower() for w in preprocessed_text] # to lower case
    preprocessed_text = [w for w in preprocessed_text if w not in stop_words]
    preprocessed_text = [w for w in preprocessed_text if len(w) > 1] # filtering short tokens
    return preprocessed_text
```

```python
print(preprocessing_pipeline(tokens))
```

```python
# Develop vocabulary
# Using all the above methods to build our nlp preprocessing


def process_docs(dir,vocab):
    for filename in listdir(dir):
        if not filename.endswith('.txt'):
            next
        directory = dir+'/'+filename
        add_doc_to_vocab(directory,vocab)
```

```python
def add_doc_to_vocab(dir,vocab):
    
    text = load_doc(dir)
    tokens = preprocessing_pipeline(text)
    vocab.update(tokens) # update counts
```

```python
from collections import Counter
vocab = Counter()
process_docs('review_polarity/txt_sentoken/neg',vocab)
process_docs('review_polarity/txt_sentoken/pos',vocab)
    
#Print the size of vocab
print(len(vocab))
```

```python
print(vocab.most_common(50))
```

```python
# Keep tokens with > 5 occurences
min_occurrence = 5

tokens = [k for k,c in vocab.items() if c>=min_occurrence]
print(len(tokens))
```

```python
# function called save list() to save a list of items, in this case, tokens to file, one per line

def save_list(lines,filename):
    text = '\n'.join(lines)
    file = open(filename,'w')
    file.write(text)
    file.close()
```

```python
save_list(tokens,'vocab.txt')
```

We can use the data cleaning and chosen vocabulary to prepare each movie review and save the
prepared versions of the reviews ready for modeling

```python
def doc_to_line(dir,vocab):
    text = load_doc(dir)
    tokens = preprocessing_pipeline(text)
    tokens = [w for w in tokens if w in vocab]
    return ' '.join(tokens)
```

```python
def process_docs(dir,vocab):
    lines = list()
    
    for filename in listdir(dir):
        if not filename.endswith('.txt'):
            next
        path = dir+'/'+filename
        line = doc_to_line(path,vocab)
        lines.append(line)
        
    return lines
```

```python
# Load vocabulary
vocab_filename = 'vocab.txt'
vocabs = load_doc(vocab_filename)
vocabs = vocabs.split()
vocabs = set(vocabs)
```

```python
#Prepare negative reviews
negative_lines = process_docs('review_polarity/txt_sentoken/neg',vocab)
print(len(negative_lines))
save_list(negative_lines,'negative.txt')
```

```python
#Prepare positive reviews 
positive_lines = process_docs('review_polarity/txt_sentoken/pos',vocab)
print(len(positive_lines))
save_list(positive_lines,'positive.txt')
```
